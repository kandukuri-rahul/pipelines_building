```python
import pandas as pd 
df1=pd.read_csv("C:\\Users\\Rahul\\Downloads\\df1_c.csv")
df2=pd.read_csv("C:\\Users\\Rahul\\Downloads\\df2_c.csv")
merged_df=pd.merge(df1,df2,on='customer_id')
```

```python
merged_df
```

```python
#captalize the names
merged_df['name'].str.title()
```

```python
#replace unknown emails with their name 
merged_df['email'].fillna(merged_df['name'].str.replace(' ','').str.lower()+'@example.com')
merged_df
```

```python
customer_df=merged_df
```

```python
customer_df['amount'].fillna(customer_df['amount'].mean())
customer_df
```

```python
customer_df['phone']=customer_df['phone'].str.replace('-','')
customer_df
```

```python
customer_df['Code']=customer_df['address'].str.slice(-8,-6)
customer_df['phone']=customer_df['phone'].astype(str)
customer_df
```

```python
def formate_phone(customer_df):
    l=len(customer_df['phone'])
    for i in range(l):
        if len(customer_df['phone'][i])>10 or len(customer_df['phone'][i])<10:
            customer_df['phone'][i]='0'
        elif customer_df['Code'][i] == 'IN':
            customer_df['phone'][i]='+91 '+str(customer_df['phone'][i])
        elif customer_df['Code'][i] == 'AE':
            customer_df['phone'][i]='+971 '+str(customer_df['phone'][i])
        elif customer_df['Code'][i] == 'FM':
            customer_df['phone'][i]='+691 '+str(customer_df['phone'][i])
        elif customer_df['Code'][i] == 'PW':
            customer_df['phone'][i]='+680 '+str(customer_df['phone'][i])
        else:
            customer_df['phone'][i]='+1 '+str(customer_df['phone'][i])  
        
    return customer_df
formate_phone(customer_df)

```

```python
customer_df.to_csv('cleaned.csv',index=False)
```
